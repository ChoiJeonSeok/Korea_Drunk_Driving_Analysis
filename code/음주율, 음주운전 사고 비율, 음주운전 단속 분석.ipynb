```python
import statsmodels.api as sm

# 독립변수(X)와 종속변수(y) 설정
X = age_all_df['음주교통사고 비율']
y = age_all_df['적발건수']

# 상수항을 추가하기 위해 X에 절편(intercept) 열을 추가
X = sm.add_constant(X)

# 선형 회귀 모델 적합
model = sm.OLS(y, X)
results = model.fit()

# 회귀분석 결과 출력
print(results.summary())
```
